```python
# importing packages
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import csv
from scipy import stats
import matplotlib.ticker as ticker
from matplotlib.patches import Polygon
# enabling pyplot charts in jupyter
%matplotlib inline

# read in the CSV
df = pd.read_csv('Project_data.csv')
df.dtypes
df
```

```python
# first I want to see if my pain was trending downward overall
# set variables for date and pain 
date = df['Date']
pain_level = df['Pain_Level']
print("What pain level rating was most common?")
# count occurrences of each pain rating from 
print(pain_level.value_counts(), '\n')

# set figure size to make it display larger
plt.figure(figsize=(8, 8))
fig, ax = plt.subplots()
plt.style.use('default')
plt.plot(date, pain_level)
# change the x axis to only plot a tick once a week for less crowding
ax.xaxis.set_major_locator(ticker.MultipleLocator(7))
plt.yticks(np.arange(5), ['0', '1', '2', '3', '4'])
ax.set_ylabel('Pain Level')
ax.set_xlabel('Date')
ax.set_title('Pain Throughout the Project')
print("There does not appear to be progress with overall pain levels from the beginning \
to the end of the project.")
```

```python
# isolate the rows where strength and cardio are listed as "Other Workouts"
# to see if this type of workout has affected my neck pain
sc = df.loc[lambda df:df['Other_Exercise'] == 'strength, cardio']
# find the indexes of those days and put them in a list
cardio_indexes = sc.index.to_list()
# make a new dataframe with only the strength cardio days
day_of_frame = df.iloc[cardio_indexes]
# new list for the day after the strength cardio days
day_after_sc = [x + 1 for x in cardio_indexes]
# make another dataframe with only the days after strength cardio
day_after_frame = df.iloc[day_after_sc]
# comparing pain on day of strength and cardio to the day after
print("Mean for day of strength & cardio:", day_of_frame['Pain_Level'].mean())
print("Mean for day after strength & cardio:", day_after_frame['Pain_Level'].mean(), '\n')
print(stats.ttest_ind(day_of_frame['Pain_Level'], day_after_frame['Pain_Level'], equal_var = False), '\n')
print('There is not a statistically significant difference in my pain \
the days I do a strength and cardio workout compared with the day after those workouts.')
```

```python
# take what I did with strength and cardio and apply to shoulder stretch workouts
stretch = df.loc[lambda df:df['Other_Exercise'] == 'stretch (shoulders, back)']
stretch_indexes = stretch.index.to_list()
day_of_stretch = df.iloc[stretch_indexes]
day_after_indexes = [x + 1 for x in stretch_indexes]
# use the iloc to get day after information
day_after_stretch = df.iloc[day_after_indexes]
# comparing pain level of day of stretch to the day after
print("Mean for day of shoulder stretches:", day_of_stretch['Pain_Level'].mean())
print("Mean for day after shoulder stretches:", day_after_stretch['Pain_Level'].mean(), '\n')
print(stats.ttest_ind(day_of_stretch['Pain_Level'], day_after_stretch['Pain_Level']), '\n')
print('The pain level on the days after shoulder stretch workouts is also not statistically significant,\
however, the p value is lower than for the day after strength and cardio workouts.')
```

```python
# evaluating whether there's a relationship between pain and hours sitting
# create a new dataframe of only days with pain of 3 or 4
# and a dataframe of only days with pain of 2 or less
high_pain_days = df.loc[lambda df:df['Pain_Level']> 2]
low_pain_days = df.loc[lambda df:df['Pain_Level']< 3]
print("Statistics of the High Pain Days:")
print(high_pain_days['Hours_Sitting'].describe(), "\n")
print("Statistics of the Low Pain Days:")
print(low_pain_days['Hours_Sitting'].describe())

# making a box plot comparing high pain and low pain days based on hours sitting
fig, axs = plt.subplots(1, 2)
plt.figure(figsize=(7, 7))
axs[0].boxplot(high_pain_days['Hours_Sitting'], widths=.7, labels=['Hours Sitting'])
axs[0].set_title('High Pain Days')
axs[1].boxplot(low_pain_days['Hours_Sitting'], widths=.7, labels=['Hours Sitting'])
axs[1].set_title('Low Pain Days')

print('\n')
print(stats.ttest_ind(high_pain_days['Hours_Sitting'], low_pain_days['Hours_Sitting'], equal_var = False))
```

```python
# comparing pain levels on days I do my therapy exercises and days I do not
yes_therapy = df.loc[lambda df:df['Therapy_Exercises'] == 'Yes']
no_therapy = df.loc[lambda df:df['Therapy_Exercises'] == 'No']
print("Days I did therapy exercises have mean pain score of", (yes_therapy['Pain_Level'].mean()))
print("Days I did not do therapy exercises have a mean pain score of", (no_therapy['Pain_Level'].mean()), '\n')
print(stats.ttest_ind(yes_therapy['Pain_Level'], no_therapy['Pain_Level'], equal_var = False))
# it appears that I am close to statistical significance with this comparison
# however, I wonder if the days I skipped the exercises are also low sitting days
```

```python
# comparing days where I do therapy exercises to days I do not
yes_sitting = yes_therapy['Hours_Sitting']
no_sitting = no_therapy['Hours_Sitting']
# statistics for each
print('Statistics for days I do therapy exercises:')
print(yes_sitting.describe(), '\n')
print('Statistics for days I do not do the exercises:')
print(no_sitting.describe())
# creating a box plot to compare
plt.figure(figsize=(7, 7))
# creating each box plot 
fig, axs = plt.subplots(1, 2)
# I used "Tufte" style to reduce data ink (showbox and showcaps).
axs[0].boxplot(yes_sitting, widths=.7, labels=['Hours Sitting'], showbox=False, showcaps=False)
axs[0].set_title('Therapy Exercises')
axs[1].boxplot(no_sitting, widths=.7, labels=['Hours Sitting'], showbox=False, showcaps=False)
axs[1].set_title('No Therapy Exercises')
# the results show that yes, the days I skipped the therapy exercises
# are also the days I spent less time sitting
```

```python
# comparing average pain per day of the week
mon = df.loc[lambda df:df['Day'] == 'M']
mon_pain = mon['Pain_Level']
tues = df.loc[lambda df:df['Day'] == 'Tu']
tues_pain = tues['Pain_Level']
wed = df.loc[lambda df:df['Day'] == 'W']
wed_pain = wed['Pain_Level']
thurs = df.loc[lambda df:df['Day'] == 'Th']
thurs_pain = thurs['Pain_Level']
fri = df.loc[lambda df:df['Day'] == 'F']
fri_pain = fri['Pain_Level']
sat = df.loc[lambda df:df['Day'] == 'Sa']
sat_pain = sat['Pain_Level']
sun = df.loc[lambda df:df['Day'] == 'Su']
sun_pain = sun['Pain_Level']
xaxis = ("Mon", "Tues", "Wed", "Thurs", "Fri", "Sat", "Sun")
days = (mon_pain.mean(), tues_pain.mean(), wed_pain.mean(), thurs_pain.mean(), fri_pain.mean(), sat_pain.mean(), sun_pain.mean())
plt.title("Average Pain Per Day of the Week")
plt.yticks(np.arange(5), ['0', '1', '2', '3', '4'])
plt.ylabel("Pain Level")
plt.xlabel("Day")
plt.bar(xaxis, days)
```

```python
mon_sitting = mon['Hours_Sitting']
tues_sitting = tues['Hours_Sitting']
wed_sitting = wed['Hours_Sitting']
thurs_sitting = thurs['Hours_Sitting']
fri_sitting = fri['Hours_Sitting']
sat_sitting = sat['Hours_Sitting']
sun_sitting = sun['Hours_Sitting']
avg_sitting = (mon_sitting.mean(), tues_sitting.mean(), wed_sitting.mean(), thurs_sitting.mean(),\
              fri_sitting.mean(), sat_sitting.mean(), sun_sitting.mean())
plt.title("Average Hours Sitting Per Day of the Week")
plt.xlabel("Day")
plt.ylabel("Hours")
plt.bar(xaxis, avg_sitting)
print("The weekend days average far fewer hours sitting than weekdays.")
```

```python
# plotting dual scales axes to see if pain and hours sitting correlate by day of week
fig, ax1 = plt.subplots()
color = 'tab:red'
ax1.set_xlabel('Days')
ax1.set_ylabel('Pain Level', color=color)
ax1.plot(xaxis, days, color=color)
ax1.tick_params(axis='y', labelcolor=color)

ax2 = ax1.twinx()  

color = 'tab:blue'
ax2.set_ylabel('Hours Sitting', color=color)
ax2.plot(xaxis, avg_sitting, color=color)
ax2.tick_params(axis='y', labelcolor=color)

fig.tight_layout()
plt.title("Pain Level and Hours Sitting by Day")
plt.show()
print("With the exception of Mondays, pain seems to correlate with hours sitting.")
```
